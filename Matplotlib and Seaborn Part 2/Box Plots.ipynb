```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv("./fuel_econ.csv")
data.head()
```

```python
data.describe()
```

Voilin plot vs Box plot

```python
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.api.types.CategoricalDtype(ordered = True, categories = sedan_classes)
data["VClass"] = data["VClass"].astype(vclasses)

plt.figure(figsize = [12, 5])

# left plot: scatterplot of discrete data with jitter and transparency
plt.subplot(1, 2, 1)
ax1 = sb.violinplot(data = data, x = 'VClass', y = 'comb');
plt.xlabel("Car Class");
plt.ylabel("Fuel efficiency");

# right plot: basic Violin Plot
plt.subplot(1, 2, 2)

sb.boxplot(data = data, x = 'VClass', y = 'comb');
plt.xlabel("Car Class");
plt.ylabel("Fuel efficiency");
plt.ylim(ax1.get_ylim()) # set y-axis limits to be the same as the left plot. 
# since the visualization range is different the box plot will be smaller and 
# will only go up to 7 or so instead of 8 on the violin plot.

```

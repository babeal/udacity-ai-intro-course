```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv("./fuel_econ.csv")
data.head()
```

```python
data.describe()
```

```python
base_color = sb.color_palette()[0]
sb.violinplot(data = data, x = "VClass", y = "comb", color = base_color, inner = "quartile")
plt.xticks(rotation = 15);
```

Create a facet grid with plots for each of the types in the VClass columns.

```python
g = sb.FacetGrid(data = data, col = "VClass");
# i tried to put g.map into the next cell but it threw an error.
# It looks like you can't have the map and the declaration of g in different cells.
g.map(plt.hist, "comb");
```

Now lets make the charts prettier with some bins and column wrap. 
sharey = False - adds Y axes to each chart. The default is true as seen above. 

```python
bins = np.arange(12, 58+2, 2)
g = sb.FacetGrid(data = data, col = "VClass", col_wrap = 3, sharey = False);
g.map(plt.hist, "comb", bins = bins);
```
